```python
import tensorflow as tf
from tensorflow.keras import layers, Model
import numpy as np
import matplotlib.pyplot as plt
import os
from glob import glob

# --- Configuración del modelo ---
IMG_HEIGHT = 128
IMG_WIDTH = 128
NUM_CLASSES = 6  # 0-background, 1-Cow-tongue, 2-Dandelion, 3-Kikuyo, 4-Other, 5-Potato
BATCH_SIZE = 16
EPOCHS = 50
LEARNING_RATE = 1e-4

# --- Paths ---
BASE_PATH = '/content/Balanced'
TRAIN_IMAGES_PATH = os.path.join(BASE_PATH, 'train/images')
TRAIN_MASKS_PATH = os.path.join(BASE_PATH, 'train/masks')
VAL_IMAGES_PATH = os.path.join(BASE_PATH, 'val/images')
VAL_MASKS_PATH = os.path.join(BASE_PATH, 'val/masks')

# --- Módulos del modelo ---
class ASPPModule(layers.Layer):
    """Atrous Spatial Pyramid Pooling Module optimizado"""
    def __init__(self, filters=64, **kwargs):
        super(ASPPModule, self).__init__(**kwargs)
        self.filters = filters

        self.conv_1x1 = layers.Conv2D(filters, 1, padding='same', use_bias=False)
        self.bn_1x1 = layers.BatchNormalization()

        self.conv_3x3_6 = layers.Conv2D(filters, 3, padding='same', dilation_rate=6, use_bias=False)
        self.bn_3x3_6 = layers.BatchNormalization()

        self.conv_3x3_12 = layers.Conv2D(filters, 3, padding='same', dilation_rate=12, use_bias=False)
        self.bn_3x3_12 = layers.BatchNormalization()

        self.global_avg_pool = layers.GlobalAveragePooling2D(keepdims=True)
        self.conv_1x1_gap = layers.Conv2D(filters, 1, padding='same', use_bias=False)
        self.bn_1x1_gap = layers.BatchNormalization()

        self.conv_final = layers.Conv2D(filters, 1, padding='same', use_bias=False)
        self.bn_final = layers.BatchNormalization()
        self.dropout = layers.Dropout(0.1)

    def call(self, inputs, training=None):
        input_shape = tf.shape(inputs)

        conv_1x1 = self.bn_1x1(self.conv_1x1(inputs), training=training)
        conv_3x3_6 = self.bn_3x3_6(self.conv_3x3_6(inputs), training=training)
        conv_3x3_12 = self.bn_3x3_12(self.conv_3x3_12(inputs), training=training)

        gap = self.global_avg_pool(inputs)
        gap = self.bn_1x1_gap(self.conv_1x1_gap(gap), training=training)
        gap = tf.image.resize(gap, [input_shape[1], input_shape[2]], method='bilinear')

        concat = layers.Concatenate()([conv_1x1, conv_3x3_6, conv_3x3_12, gap])

        output = self.bn_final(self.conv_final(concat), training=training)
        output = self.dropout(output, training=training)

        return output

class DeformableAttention(layers.Layer):
    """Simplified Deformable Attention Layer"""
    def __init__(self, filters, **kwargs):
        super(DeformableAttention, self).__init__(**kwargs)
        self.filters = filters

    def build(self, input_shape):
        self.attention_conv = layers.Conv2D(self.filters, 1, padding='same', activation='sigmoid', name='attention_weights_conv', use_bias=False)
        self.bn_attention = layers.BatchNormalization()

        self.feature_conv = layers.Conv2D(self.filters, 3, padding='same', activation='relu', name='feature_processing_conv', use_bias=False)
        self.bn_feature = layers.BatchNormalization()

        super(DeformableAttention, self).build(input_shape)

    def call(self, inputs, training=None):
        attention_weights = self.bn_attention(self.attention_conv(inputs), training=training)
        features = self.bn_feature(self.feature_conv(inputs), training=training)
        attended_features = features * attention_weights
        return attended_features


def create_efficient_backbone(input_shape):
    """Create EfficientNetV2-B0 backbone for feature extraction with pre-trained weights."""
    base_model = tf.keras.applications.EfficientNetV2B0(
        input_shape=input_shape,
        weights='imagenet',  # Cargar pesos preentrenados
        include_top=False
    )

    layer_names = [
        'block1a_project_activation',
        'block2a_expand_activation',
        'block3a_expand_activation',
        'block4a_expand_activation',
        'top_activation'
    ]

    outputs = [base_model.get_layer(name).output for name in layer_names]

    return Model(inputs=base_model.input, outputs=outputs, name='EfficientNetV2B0_Backbone')

def create_weed_detection_model(input_shape=(IMG_HEIGHT, IMG_WIDTH, 3), num_classes=NUM_CLASSES):
    """Create the complete weed detection (segmentation) model with more skip connections."""
    inputs = layers.Input(shape=input_shape)

    backbone = create_efficient_backbone(input_shape)
    backbone.trainable = False
    backbone_features = backbone(inputs)

    deepest_features = backbone_features[-1] # Shape ~ (None, 4, 4, 1280)
    block4_features = backbone_features[-2] # Shape ~ (None, 16, 16, 192) - Based on your error, not 8x8!
    block3_features = backbone_features[-3] # Shape ~ (None, 16, 16, 128) - This might be 32x32. We need to check
    block2_features = backbone_features[-4] # Shape ~ (None, 32, 32, 64) - This might be 64x64
    block1_features = backbone_features[-5] # Shape ~ (None, 64, 64, 16) - This might be 128x128/input size

    # Para depurar las formas reales de las salidas del backbone,
    # puedes descomentar estas líneas y ejecutar el modelo una vez:
    # input_tensor = tf.zeros((1, IMG_HEIGHT, IMG_WIDTH, 3))
    # output_shapes = [f.shape for f in backbone(input_tensor)]
    # print(f"EfficientNetV2B0 backbone output shapes: {output_shapes}")
    # (None, 64, 64, 16), (None, 32, 32, 64), (None, 16, 16, 128), (None, 16, 16, 192), (None, 4, 4, 1280)

    # Basado en tu error anterior:
    # block4_features (originalmente (None, 16, 16, 192))
    # conv2d_transpose_3 output (decoder_lvl1) (None, 8, 8, 64)

    # --- Decodificador con redimensionamiento explícito ---

    aspp_features = ASPPModule(filters=64)(deepest_features)
    attended_features = DeformableAttention(filters=64)(aspp_features)

    # Nivel 1: Upsample desde attended_features (4x4) a 8x8 (para coincidir con el target)
    decoder_lvl1 = layers.Conv2DTranspose(64, 2, strides=2, padding='same', use_bias=False)(attended_features)
    decoder_lvl1 = layers.BatchNormalization()(decoder_lvl1)

    # Redimensionar block4_features (que es 16x16) a 8x8 para que coincida con decoder_lvl1
    block4_features_resized = layers.Resizing(
        height=8,
        width=8,
        interpolation='bilinear'
    )(block4_features)

    block4_processed = layers.Conv2D(64, 1, padding='same', use_bias=False)(block4_features_resized)
    block4_processed = layers.BatchNormalization()(block4_processed)
    decoder_lvl1_concat = layers.Concatenate()([decoder_lvl1, block4_processed])
    decoder_lvl1_conv = layers.Conv2D(64, 3, padding='same', activation='relu', use_bias=False)(decoder_lvl1_concat)
    decoder_lvl1_conv = layers.BatchNormalization()(decoder_lvl1_conv)


    # Nivel 2: Upsample de 8x8 (decoder_lvl1_conv) a 16x16
    decoder_lvl2 = layers.Conv2DTranspose(64, 2, strides=2, padding='same', use_bias=False)(decoder_lvl1_conv)
    decoder_lvl2 = layers.BatchNormalization()(decoder_lvl2)

    # Redimensionar block3_features (que según tu summary anterior podría ser 16x16) a 16x16.
    # Si block3_features también fuera 16x16, esta capa de Resizing no cambiará el tamaño, pero asegura la forma estática.
    # Si block3_features es, por ejemplo, 32x32, esto lo reducirá.
    block3_features_resized = layers.Resizing(
        height=16,
        width=16,
        interpolation='bilinear'
    )(block3_features)

    block3_processed = layers.Conv2D(64, 1, padding='same', use_bias=False)(block3_features_resized)
    block3_processed = layers.BatchNormalization()(block3_processed)
    decoder_lvl2_concat = layers.Concatenate()([decoder_lvl2, block3_processed])
    decoder_lvl2_conv = layers.Conv2D(64, 3, padding='same', activation='relu', use_bias=False)(decoder_lvl2_concat)
    decoder_lvl2_conv = layers.BatchNormalization()(decoder_lvl2_conv)

    # Nivel 3: Upsample de 16x16 (decoder_lvl2_conv) a 32x32
    decoder_lvl3 = layers.Conv2DTranspose(32, 2, strides=2, padding='same', use_bias=False)(decoder_lvl2_conv)
    decoder_lvl3 = layers.BatchNormalization()(decoder_lvl3)

    # Redimensionar block2_features (que según tu summary anterior podría ser 32x32) a 32x32.
    block2_features_resized = layers.Resizing(
        height=32,
        width=32,
        interpolation='bilinear'
    )(block2_features)

    block2_processed = layers.Conv2D(32, 1, padding='same', use_bias=False)(block2_features_resized)
    block2_processed = layers.BatchNormalization()(block2_processed)
    decoder_lvl3_concat = layers.Concatenate()([decoder_lvl3, block2_processed])
    decoder_lvl3_conv = layers.Conv2D(32, 3, padding='same', activation='relu', use_bias=False)(decoder_lvl3_concat)
    decoder_lvl3_conv = layers.BatchNormalization()(decoder_lvl3_conv)

    # Nivel 4: Upsample de 32x32 (decoder_lvl3_conv) a 64x64
    decoder_lvl4 = layers.Conv2DTranspose(16, 2, strides=2, padding='same', use_bias=False)(decoder_lvl3_conv)
    decoder_lvl4 = layers.BatchNormalization()(decoder_lvl4)

    # Redimensionar block1_features (que es 64x64) a 64x64.
    block1_features_resized = layers.Resizing(
        height=64,
        width=64,
        interpolation='bilinear'
    )(block1_features)

    block1_processed = layers.Conv2D(16, 1, padding='same', use_bias=False)(block1_features_resized)
    block1_processed = layers.BatchNormalization()(block1_processed)
    decoder_lvl4_concat = layers.Concatenate()([decoder_lvl4, block1_processed])
    decoder_lvl4_conv = layers.Conv2D(16, 3, padding='same', activation='relu', use_bias=False)(decoder_lvl4_concat)
    decoder_lvl4_conv = layers.BatchNormalization()(decoder_lvl4_conv)

    # Nivel final: Upsample de 64x64 a 128x128
    final_upsample = layers.Conv2DTranspose(num_classes, 2, strides=2, padding='same', use_bias=False)(decoder_lvl4_conv)
    final_upsample = layers.BatchNormalization()(final_upsample)

    outputs = layers.Conv2D(num_classes, 1, padding='same', activation='softmax', name='segmentation_output')(final_upsample)

    model = Model(inputs=inputs, outputs=outputs, name='WeedDetectionModel')

    return model

# --- Métricas y Funciones de Pérdida ---
def dice_coefficient(y_true, y_pred, smooth=1e-6):
    """Dice coefficient for multiclass segmentation, averaged over classes"""
    y_true_f = tf.cast(tf.reshape(y_true, [-1, NUM_CLASSES]), tf.float32)
    y_pred_f = tf.cast(tf.reshape(y_pred, [-1, NUM_CLASSES]), tf.float32)

    intersection = tf.reduce_sum(y_true_f * y_pred_f, axis=0)
    dice_scores = (2. * intersection + smooth) / (tf.reduce_sum(y_true_f, axis=0) + tf.reduce_sum(y_pred_f, axis=0) + smooth)

    return tf.reduce_mean(dice_scores)

def dice_loss(y_true, y_pred):
    """Dice loss function (1 - Dice Coefficient)"""
    return 1 - dice_coefficient(y_true, y_pred)

def combined_loss(y_true, y_pred):
    """Combined loss: Categorical Crossentropy + Dice Loss"""
    ce_loss = tf.keras.losses.categorical_crossentropy(y_true, y_pred)
    d_loss = dice_loss(y_true, y_pred)
    return ce_loss + d_loss

def iou_metric(y_true, y_pred):
    """IoU (Intersection over Union) metric for multiclass segmentation, averaged over classes"""
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(y_pred, tf.float32)

    y_pred_argmax = tf.argmax(y_pred, axis=-1)
    y_pred_one_hot = tf.one_hot(y_pred_argmax, depth=NUM_CLASSES)

    intersection = tf.reduce_sum(y_true * y_pred_one_hot, axis=[1, 2, 3])
    union = tf.reduce_sum(y_true, axis=[1, 2, 3]) + tf.reduce_sum(y_pred_one_hot, axis=[1, 2, 3]) - intersection

    iou = tf.where(tf.equal(union, 0), 1.0, intersection / union)
    return tf.reduce_mean(iou)

def iou_per_class(y_true, y_pred):
    """IoU (Intersection over Union) metric for multiclass segmentation, per class"""
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(y_pred, tf.float32)

    y_pred_argmax = tf.argmax(y_pred, axis=-1)
    y_pred_one_hot = tf.one_hot(y_pred_argmax, depth=NUM_CLASSES)

    intersection = tf.reduce_sum(y_true * y_pred_one_hot, axis=[1, 2])
    union = tf.reduce_sum(y_true, axis=[1, 2]) + tf.reduce_sum(y_pred_one_hot, axis=[1, 2]) - intersection

    iou = tf.where(tf.equal(union, 0), 1.0, intersection / union)
    return tf.reduce_mean(iou, axis=0)

# --- Carga y Preprocesamiento de Datos ---
def load_and_preprocess_data(images_path, masks_path):
    """Load and preprocess the dataset from specified paths"""
    def load_image(image_path):
        image = tf.io.read_file(image_path)
        image = tf.image.decode_image(image, channels=3, expand_animations=False)
        image.set_shape([None, None, 3])
        image = tf.image.resize(image, [IMG_HEIGHT, IMG_WIDTH])
        image = tf.cast(image, tf.float32) / 255.0
        return image

    def load_mask(mask_path):
        mask = tf.io.read_file(mask_path)
        mask = tf.image.decode_png(mask, channels=1)
        mask = tf.image.resize(mask, [IMG_HEIGHT, IMG_WIDTH], method='nearest')
        mask = tf.cast(mask, tf.int32)
        mask = tf.squeeze(mask, axis=-1)
        mask = tf.one_hot(mask, NUM_CLASSES)
        return mask

    image_paths = sorted(glob(os.path.join(images_path, '*.jpg')))
    mask_paths = sorted(glob(os.path.join(masks_path, '*.png')))

    if not image_paths:
        raise FileNotFoundError(f"No image files found in: {images_path}. Check path and file extensions.")
    if not mask_paths:
        raise FileNotFoundError(f"No mask files found in: {masks_path}. Check path and file extensions.")

    if len(image_paths) != len(mask_paths):
        print(f"DEBUG: Found {len(image_paths)} images and {len(mask_paths)} masks.")
        print("DEBUG: First 5 image files:", [os.path.basename(p) for p in image_paths[:5]])
        print("DEBUG: First 5 mask files:", [os.path.basename(p) for p in mask_paths[:5]])
        raise ValueError("Number of images and masks do not match. Check filenames and ensure a 1-to-1 correspondence.")

    print(f"Found {len(image_paths)} images and {len(mask_paths)} masks")

    image_ds = tf.data.Dataset.from_tensor_slices(image_paths).map(load_image, num_parallel_calls=tf.data.AUTOTUNE)
    mask_ds = tf.data.Dataset.from_tensor_slices(mask_paths).map(load_mask, num_parallel_calls=tf.data.AUTOTUNE)

    dataset = tf.data.Dataset.zip((image_ds, mask_ds))

    return dataset

def augment_data(image, mask):
    """Data augmentation"""
    if tf.random.uniform([]) > 0.5:
        k = tf.random.uniform([], maxval=4, dtype=tf.int32)
        image = tf.image.rot90(image, k)
        mask = tf.image.rot90(mask, k)

    if tf.random.uniform([]) > 0.5:
        image = tf.image.flip_left_right(image)
        mask = tf.image.flip_left_right(mask)

    if tf.random.uniform([]) > 0.5:
        image = tf.image.flip_up_down(image)
        mask = tf.image.flip_up_down(mask)

    if tf.random.uniform([]) > 0.5:
        min_crop_size = int(IMG_HEIGHT * 0.8)
        max_crop_size = IMG_HEIGHT
        crop_size_h = tf.random.uniform([], minval=min_crop_size, maxval=max_crop_size + 1, dtype=tf.int32)
        crop_size_w = tf.random.uniform([], minval=min_crop_size, maxval=max_crop_size + 1, dtype=tf.int32)

        combined = tf.concat([image, mask], axis=-1)
        cropped_combined = tf.image.random_crop(combined, size=[crop_size_h, crop_size_w, 3 + NUM_CLASSES])

        image = tf.image.resize(cropped_combined[..., :3], [IMG_HEIGHT, IMG_WIDTH])
        mask = tf.image.resize(cropped_combined[..., 3:], [IMG_HEIGHT, IMG_WIDTH], method='nearest')

    image = tf.image.random_brightness(image, max_delta=0.2)
    image = tf.image.random_contrast(image, lower=0.8, upper=1.2)
    image = tf.image.random_saturation(image, lower=0.8, upper=1.2)
    image = tf.image.random_hue(image, max_delta=0.1)

    image = tf.clip_by_value(image, 0.0, 1.0)

    return image, mask

def create_datasets():
    """Create training and validation datasets with augmentation and batching"""
    train_dataset = load_and_preprocess_data(TRAIN_IMAGES_PATH, TRAIN_MASKS_PATH)
    val_dataset = load_and_preprocess_data(VAL_IMAGES_PATH, VAL_MASKS_PATH)

    train_dataset = train_dataset.shuffle(buffer_size=1000)
    val_dataset = val_dataset.shuffle(buffer_size=1000)

    train_dataset = train_dataset.map(augment_data, num_parallel_calls=tf.data.AUTOTUNE)
    train_dataset = train_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
    val_dataset = val_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

    print(f"Training batches (approx): {len(list(train_dataset)) if tf.data.experimental.cardinality(train_dataset) != tf.data.experimental.INFINITE_CARDINALITY else 'unknown'}")
    print(f"Validation batches (approx): {len(list(val_dataset)) if tf.data.experimental.cardinality(val_dataset) != tf.data.experimental.INFINITE_CARDINALITY else 'unknown'}")

    return train_dataset, val_dataset

def train_model(train_dataset, val_dataset, model_save_path='best_weed_model.h5'):
    """Train the weed detection model with specified callbacks"""
    model = create_weed_detection_model()
    model.summary()

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE, clipnorm=1.0),
        loss=combined_loss,
        metrics=[dice_coefficient, iou_metric, 'accuracy']
    )

    callbacks = [
        tf.keras.callbacks.ModelCheckpoint(
            model_save_path,
            monitor='val_dice_coefficient',
            save_best_only=True,
            mode='max',
            verbose=1
        ),
        tf.keras.callbacks.EarlyStopping(
            monitor='val_dice_coefficient',
            patience=25,
            mode='max',
            restore_best_weights=True
        ),
        tf.keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.2,
            patience=8,
            min_lr=1e-7,
            verbose=1
        )
    ]

    print(f"Starting training for {EPOCHS} epochs...")
    history = model.fit(
        train_dataset,
        epochs=EPOCHS,
        validation_data=val_dataset,
        callbacks=callbacks,
        verbose=1
    )

    return model, history

def visualize_predictions(model, val_dataset, num_samples=4):
    """Visualize original images, true masks, and predicted masks"""
    class_names = ['Background', 'Cow-tongue', 'Dandelion', 'Kikuyo', 'Other', 'Potato']

    print(f"Visualizing {num_samples} samples from the validation set...")

    for images, masks in val_dataset.take(1):
        predictions = model.predict(images)

        fig, axes = plt.subplots(num_samples, 3, figsize=(15, 5 * num_samples))

        for i in range(min(num_samples, images.shape[0])):
            axes[i, 0].imshow(images[i])
            axes[i, 0].set_title('Original Image')
            axes[i, 0].axis('off')

            true_mask = np.argmax(masks[i].numpy(), axis=-1)
            axes[i, 1].imshow(true_mask, cmap='tab10', vmin=0, vmax=NUM_CLASSES - 1)
            axes[i, 1].set_title('True Mask')
            axes[i, 1].axis('off')

            pred_mask = np.argmax(predictions[i], axis=-1)
            axes[i, 2].imshow(pred_mask, cmap='tab10', vmin=0, vmax=NUM_CLASSES - 1)
            axes[i, 2].set_title('Predicted Mask')
            axes[i, 2].axis('off')

        plt.tight_layout()
        plt.show()
        break

# --- Ejecución principal ---
if __name__ == "__main__":
    if not os.path.exists(TRAIN_IMAGES_PATH):
        print(f"Error: La ruta de imágenes de entrenamiento NO existe: {TRAIN_IMAGES_PATH}")
        exit()
    if not os.path.exists(TRAIN_MASKS_PATH):
        print(f"Error: La ruta de máscaras de entrenamiento NO existe: {TRAIN_MASKS_PATH}")
        exit()
    if not os.path.exists(VAL_IMAGES_PATH):
        print(f"Error: La ruta de imágenes de validación NO existe: {VAL_IMAGES_PATH}")
        exit()
    if not os.path.exists(VAL_MASKS_PATH):
        print(f"Error: La ruta de máscaras de validación NO existe: {VAL_MASKS_PATH}")
        exit()

    print("Creando datasets...")
    train_ds, val_ds = create_datasets()

    print("Entrenando modelo...")
    model, history = train_model(train_ds, val_ds)

    print("Visualizando predicciones...")
    visualize_predictions(model, val_ds)

    print("¡Entrenamiento completado! El mejor modelo se guardó como 'best_weed_model.h5'")

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['dice_coefficient'], label='Dice Coeff (Train)')
    plt.plot(history.history['val_dice_coefficient'], label='Dice Coeff (Validation)')
    plt.title('Dice Coefficient over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Dice Coefficient')
    plt.legend()
    plt.grid(True)

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Loss (Train)')
    plt.plot(history.history['val_loss'], label='Loss (Validation)')
    plt.title('Loss over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    plt.show()

    print("\nEvaluando el modelo final en el conjunto de validación:")
    final_metrics = model.evaluate(val_ds)
    print(f"Validation Loss: {final_metrics[0]:.4f}")
    print(f"Validation Dice Coefficient: {final_metrics[1]:.4f}")
    print(f"Validation IoU: {final_metrics[2]:.4f}")
    print(f"Validation Accuracy: {final_metrics[3]:.4f}")

    # Calculate IoU per class
    iou_per_class_values = []
    for images, masks in val_ds:
        predictions = model.predict(images)
        iou_per_class_values.append(iou_per_class(masks, predictions))

    iou_per_class_values = np.mean(iou_per_class_values, axis=0)
    class_names = ['Background', 'Cow-tongue', 'Dandelion', 'Kikuyo', 'Other', 'Potato']
    print("\nIoU por clase:")
    for i, iou in enumerate(iou_per_class_values):
        print(f"  IoU para {class_names[i]}: {iou:.4f}")
```
